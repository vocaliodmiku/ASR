````
#build a large lexicon that invovles words in both the training and decoding.
(
  echo "make word graph ..."

  cd $H; mkdir -p data/{dict,lang,graph} && \
  cp $thchs/resource/dict/{extra_questions.txt,nonsilence_phones.txt,optional_silence.txt,silence_phones.txt} data/dict && \
  # 将下载到的lexicon.txt复制到运行路径数据目录中
  cat $thchs/resource/dict/lexicon.txt $thchs/data_thchs30/lm_word/lexicon.txt | \
  # 反向查找去除开始结束符号并排序
  grep -v '<s>' | grep -v '</s>' | sort -u > data/dict/lexicon.txt || exit 1;
  # prepare_lang 这里可以参考yesno tutorial中
  # 1.3   utils/prepare_lang.sh 部分. 这个脚本会完成对data/lang目录的准备。
  # 给定lexicon.txt,其格式为 word phone1 phone2 ... phoneN 或者说是lexiconp.txt
  # 其格式为 word pron_prob phone1 phone2 ... phoneN（优先使用lexiconp.txt）
  # 以及刚才复制过来的那些文件
  # nonsilence_phones.txt：real phone 也即非静音音素。
  # silence_phones.txt：包括不同类型的噪声笑声咳嗽
  # 这些文件每一行的众多音素，都被认为是由同一个基础音素变化重音和语调而来
  # optional_silence.txt：只包括SIL。extra_questions.txt：为空
  # 比较重要的是生成的lexicon FST （L.fst）
  utils/prepare_lang.sh --position_dependent_phones false data/dict "<SPOKEN_NOISE>" data/local/lang data/lang || exit 1;
  # 压缩3-gram模型
  gzip -c $thchs/data_thchs30/lm_word/word.3gram.lm > data/graph/word.3gram.lm.gz || exit 1;
  # 将ARPA格式的语言模型转换为FST（G.fst）
  utils/format_lm.sh data/lang data/graph/word.3gram.lm.gz $thchs/data_thchs30/lm_word/lexicon.txt data/graph/lang || exit 1;
)

#make_phone_graph
( # 再放送（音素版本的）。这两个一个在data/lang中，一个data/lang_phone中
  echo "make phone graph ..."
  cd $H; mkdir -p data/{dict_phone,graph_phone,lang_phone} && \
  cp $thchs/resource/dict/{extra_questions.txt,nonsilence_phones.txt,optional_silence.txt,silence_phones.txt} data/dict_phone  && \
  cat $thchs/data_thchs30/lm_phone/lexicon.txt | grep -v '<eps>' | sort -u > data/dict_phone/lexicon.txt  && \
  echo "<SPOKEN_NOISE> sil " >> data/dict_phone/lexicon.txt  || exit 1;
  utils/prepare_lang.sh --position_dependent_phones false data/dict_phone "<SPOKEN_NOISE>" data/local/lang_phone data/lang_phone || exit 1;
  gzip -c $thchs/data_thchs30/lm_phone/phone.3gram.lm > data/graph_phone/phone.3gram.lm.gz  || exit 1;
  utils/format_lm.sh data/lang_phone data/graph_phone/phone.3gram.lm.gz $thchs/data_thchs30/lm_phone/lexicon.txt \
    data/graph_phone/lang  || exit 1;
)
````

**3. 构建模型** 

![论文展示的结构](https://www.researchgate.net/profile/Dong_Wang19/publication/286301812/figure/fig1/AS:391381437698048@1470324014196/Architecture-of-the-THCHS-30-baseline-sys-tem.png)  

论文地址 ： http://arxiv.org/abs/1512.01882  

以下我复读一遍其模型设置：  

    我们使用Kaldi工具来训练HMM-DNN混合音学模型，其结构如上图所示。首先会使用MFCC特征及其一二阶差分系数训练一个单音素GMM模型。进行倒谱均值归一化是为了减轻通道效应（环境不匹配杂音）。在这个单音素的对齐基础上继续使用经过LDA和MLLT转换过的特征进行三音素的模型构建，其中LDA是Linear Discriminant Analysis 是有监督的降维方法，MLLT是Maximum Likelihood Linear Transform也就是最大似然线性变换。接着对模型进行说话人自适应改造，至此GMM将会为后面的DNN提供对齐状态。
    基于GMM提供的对齐数据，接着进行DNN的训练，特征是40维的Fbanks特征，相邻的帧通过一个帧长为11的窗进行连接，也即是左右5个帧。连接后的特征经过LDA降维，减少到200维。然后归整数据为一个全局的期望和方差来得到DNN模型的输入。DNN的由4个隐含层组成，每个隐含层包括1200个单元。输出层由3386个单元组成。

thchs30中构建模型部分调用了许多模块化的方法，这里可以先参考下个代码框的注解，然后对应模型构建代码。流程与上面说的模型一致，基本上每三行为一个部分（训练解码与对齐）：

不同位置上的对齐也稍微有点区别。在SAT之后，使用适用于说话人自适应的fMLLR特征变换。

````
# train_mono.sh
# 单音素训练脚本：

//初始化，[topo feats] -> [0.mdl tree]
gmm-init-mono 
//生成训练图，[0.mdl text l.fst] -> [train.fst]
compile-train-graph
//对标签进行初始化对齐[train.fst feats 0.mdl tree] -> [1.ali]
align-equal-compiled
//统计估计模型所需统计量，[feats 1.ali] -> [1.acc]
gmm-acc-stats-ali
//参数重估，估计新的模型 [1.acc] -> [1.mdl]
gmm-est

//迭代训练
for i < iteration
    //重新统计所需统计量，[$i.ali] -> [$i.acc]
    gmm-acc-stats-ali
    //估计新的模型，[$i.acc $i.mdl] -> [$i+1.mdl]
    gmm-est

//输出最后的模型
final.mdl = $i.mdl
# 单独的对齐脚本 e.g align_si.sh
//重新对齐，[train.fst $i.mdl] ->[$i+1.ali] 
gmm-align-compiled


# train_deltas.sh
# 三音素训练脚本：

//特征处理 [feats] -> [feats]
apply-cmvn | add-deltas
//由生成的单音素模型的对齐结果对三音素参数统计，用于生成决策树[final.ali feats] -> [treeacc]
acc-tree-stats
//三音素绑定，[treeacc] -> [tree]
cluster-phone
compile-questions
build-tree //该步骤完成决策树三音素聚类
//三音素模型初始化，[treeacc tree topo] -> [1.occ 1.mdl] -> [1.mdl]
gmm-init-model | gmm-mixup
//将单音素对齐文件中的元素替换为决策树的叶子，[final.mdl 1.mdl final.ali] -> [ali.new]
# 这个函数体现出了为什么之前使用transition-id 而不是 pdfs-id
convert-ali 
//生成训练图，[1.mdl text l.fst] -> [train.fst]
compile-train-graph

//迭代训练
for i < iteration
    //重新对齐，[train.fst $i.mdl] ->[$i+1.ali] 
    gmm-align-compiled
    //重新统计所需统计量，[$i.ali] -> [$i.acc]
    gmm-acc-stats-ali
    //估计新的模型，[$i.acc] -> [$i.mdl]
    gmm-est //该步骤增加混合高斯分量的数目
//输出最后的模型
final.mdl = $i.mdl

# train_lda_mllt.sh
# lda-mllt训练脚本，非说话人自适应，mllt的作用是减少协方差矩阵对角化的损失：

//生成先验概率，统计计算lda所需统计量，[splice-feats final.ali] -> [lda.acc]
ali-to-post
weight-silence-post
acc-lda
//估计lda矩阵，[lda.acc] -> [lda.mat]
est-lda
//通过对转换后的特征重新统计，用于生成决策树[final.ali feats.*lda.mat] -> [treeacc]
acc-tree-stats
//三音素绑定，[treeacc] -> [tree]
cluster-phone
compile-questions
build-tree //该步骤完成决策树三音素聚类
//三音素模型初始化，[treeacc tree topo] -> [1.occ 1.mdl]
gmm-init-model/gmm-init-model-flat
//将三音素决策树/ 模型的对齐文件转换为其他树/模型的对齐文件，[final.mdl 1.mdl final.ali] -> [ali.new]
convert-ali 
//生成训练图，[1.mdl text l.fst] -> [train.fst]
compile-train-graph

//迭代训练
for i < iteration
    //重新对齐，[train.fst $i.mdl] ->[$i+1.ali] 
    gmm-align-compiled
        //同lda，估计mllt的矩阵
        ali-to-post | weight-silence-post | gmm-acc-mllt
        est-mllt
        //对gmm模型进行变换，[mllt.mat mdl] -> [new.mdl]
        gmm-transform-means
        //组合变换矩阵，[lda.mat mllt.mat] -> [lda.mat]
        compose-transforms
    //重新统计所需统计量，[$i.ali] -> [$i.acc]
    gmm-acc-stats-ali
    //估计新的模型，[$i.acc] -> [$i.mdl]
    gmm-est //该步骤增加混合高斯分量的数目
//输出最后的模型
final.mdl = $i.mdl

# train_sat.sh
# 说话人自适应模型，fmllr训练脚本：

//生成先验概率，统计计算fmllr所需统计量，[splice-feats spk2utt] -> [trans]
ali-to-post
weight-silence-post
gmm-est-fmllr
//通过对转换后的特征重新统计，用于生成决策树[final.ali feats.*lda.mat] -> [treeacc]
acc-tree-stats
//三音素绑定，[treeacc] -> [tree]
cluster-phone
compile-questions
build-tree //该步骤完成决策树三音素聚类
//三音素模型初始化，[treeacc tree topo] -> [1.occ 1.mdl]
gmm-init-model
//将三音素决策树的叶子替换为转换后模型决策树的叶子，[final.mdl 1.mdl final.ali] -> [ali.new]
convert-ali 
//生成训练图，[1.mdl text l.fst] -> [train.fst]
compile-train-graph

//迭代训练
for i < iteration
    //重新对齐，[train.fst $i.mdl] ->[$i+1.ali] 
    gmm-align-compiled
        //同lda，估计fmllr的矩阵 -> [fmllr.trans]
        ali-to-post | weight-silence-post | gmm-est-fmllr
        //组合变换矩阵，[trans.mat fmllr.trans] -> [trans.mat]
        compose-transforms
    //重新统计所需统计量，[$i.ali] -> [$i.acc]
    gmm-acc-stats-ali
    //估计新的模型，[$i.acc] -> [$i.mdl]
    gmm-est //该步骤增加混合高斯分量的数目
//输出最后的模型
final.mdl = $i.mdl
# train_quick.sh
# 用于在现有特征上训练模型。对于当前模型中在树构建之后的每个状态，基于树统计量的重叠数目相似度来选择旧模型中最接近的状态。
check_phones_compatible.sh
//生成先验概率，统计计算lda所需统计量，[splice-feats final.ali] -> [lda.acc]
ali-to-post
//估计lda矩阵，[lda.acc] -> [lda.mat]
est-lda
//通过对转换后的特征重新统计，用于生成决策树[final.ali feats.*lda.mat] -> [treeacc]
acc-tree-stats
//对统计量求和
sum-tree-stats
//三音素绑定，[treeacc] -> [tree]
cluster-phones
compile-questions
build-tree //该步骤完成决策树三音素聚类
//三音素模型初始化，[treeacc tree topo] -> [1.occ 1.mdl]
gmm-init-model
//将三音素决策树/模型的对齐文件替换为转换后模型决策树/模型的对齐文件，[final.mdl 1.mdl final.ali] -> [ali.new]
convert-ali
compile-train-graphs

//迭代训练
for i < iteration
    gmm-align-compiled
    gmm-acc-stats-ali
    gmm-est
// 分析对齐数据
analyze_alignments.sh

# 参考自一篇博客https://www.cnblogs.com/sunhongwen/p/9507252.html
````

````
#run_dnn.sh
#在看这个脚本之前建议看文档，即使这里会添加注释。kaldi的神经网络实现跟tf,keras之类的不太一样
#http://www.kaldi-asr.org/doc/dnn.html

#run from ../..
#DNN training, both xent and MPE


. ./cmd.sh ## You'll want to change cmd.sh to something that will work on your system.
           ## This relates to the queue.

. ./path.sh ## Source the tools/utils (import the queue.pl)

stage=0
nj=8

. utils/parse_options.sh || exit 1;
# gmmdir=exp/tri4b
# alidir=exp/tri4b_ali
# alidir_cv=exp/tri4b_ali_cv
gmmdir=$1
alidir=$2
alidir_cv=$3
# 第一步依然是对原始输入进行特征抽取。在模型结构图中，音频频谱图作为输入，这里将提取它的fbank
# 来作为特征，当然还有倒谱均值方差归一化的操作。
#generate fbanks
if [ $stage -le 0 ]; then
  echo "DNN training: stage 0: feature generation"
  rm -rf data/fbank && mkdir -p data/fbank &&  cp -R data/{train,dev,test,test_phone} data/fbank || exit 1;
  for x in train dev test; do
    echo "producing fbank for $x"
    #fbank generation
    steps/make_fbank.sh --nj $nj --cmd "$train_cmd" data/fbank/$x exp/make_fbank/$x fbank/$x || exit 1
    #ompute cmvn
    steps/compute_cmvn_stats.sh data/fbank/$x exp/fbank_cmvn/$x fbank/$x || exit 1
  done

  echo "producing test_fbank_phone"
  cp data/fbank/test/feats.scp data/fbank/test_phone && cp data/fbank/test/cmvn.scp data/fbank/test_phone || exit 1;

fi


#xEnt training
if [ $stage -le 1 ]; then
  outdir=exp/tri4b_dnn
  # 帧交叉熵训练，train.sh参数中指定了网络的结构
  # hid-layer=4；hid_dim=1024；lr=0.008
  # train.sh <data-train> <data-dev> <lang-dir> <ali-train> <ali-dev> <exp-dir>
  #NN training
  (tail --pid=$$ -F $outdir/log/train_nnet.log 2>/dev/null)& # forward log
  $cuda_cmd $outdir/log/train_nnet.log \
    steps/nnet/train.sh --copy_feats false --cmvn-opts "--norm-means=true --norm-vars=false" --hid-layers 4 --hid-dim 1024 \
    --learn-rate 0.008 data/fbank/train data/fbank/dev data/lang $alidir $alidir_cv $outdir || exit 1;
  #Decode (reuse HCLG graph in gmmdir)
  #Usage: $0 [options] <graph-dir> <data-dir> <decode-dir>
  #这里word与phone各自都会解码
  (
    steps/nnet/decode.sh --nj $nj --cmd "$decode_cmd" --srcdir $outdir --config conf/decode_dnn.config --acwt 0.1 \
      $gmmdir/graph_word data/fbank/test $outdir/decode_test_word || exit 1;
  )&
  (
   steps/nnet/decode.sh --nj $nj --cmd "$decode_cmd" --srcdir $outdir --config conf/decode_dnn.config --acwt 0.1 \
     $gmmdir/graph_phone data/fbank/test_phone $outdir/decode_test_phone || exit 1;
  )&

fi

#MPE training
# sMBR序列区分性训练，这个阶段的训练是联合优化整个句子，比起帧级别训练，显然这个要更接近ASR的目标
srcdir=exp/tri4b_dnn
acwt=0.1

if [ $stage -le 2 ]; then
  # generate lattices and alignments
  steps/nnet/align.sh --nj $nj --cmd "$train_cmd" \
    data/fbank/train data/lang $srcdir ${srcdir}_ali || exit 1;
  steps/nnet/make_denlats.sh --nj $nj --cmd "$decode_cmd" --config conf/decode_dnn.config --acwt $acwt \
    data/fbank/train data/lang $srcdir ${srcdir}_denlats || exit 1;
fi

if [ $stage -le 3 ]; then
  outdir=exp/tri4b_dnn_mpe
  #Re-train the DNN by 3 iteration of MPE
  steps/nnet/train_mpe.sh --cmd "$cuda_cmd" --num-iters 3 --acwt $acwt --do-smbr false \
    data/fbank/train data/lang $srcdir ${srcdir}_ali ${srcdir}_denlats $outdir || exit 1
  #Decode (reuse HCLG graph)
  for ITER in 3 2 1; do
   (
    steps/nnet/decode.sh --nj $nj --cmd "$decode_cmd" --nnet $outdir/${ITER}.nnet --config conf/decode_dnn.config --acwt $acwt \
      $gmmdir/graph_word data/fbank/test $outdir/decode_test_word_it${ITER} || exit 1;
   )&
   (
   steps/nnet/decode.sh --nj $nj --cmd "$decode_cmd" --nnet $outdir/${ITER}.nnet --config conf/decode_dnn.config --acwt $acwt \
     $gmmdir/graph_phone data/fbank/test_phone $outdir/decode_test_phone_it${ITER} || exit 1;
   )&
  done
fi
````
